```python
from itertools import combinations
from collections import Counter
```

```python
class HandType():
    HIGH_CARD = 1
    PAIR = 2
    TWO_PAIR = 3
    THREE_OF_A_KIND = 4
    STRAIGHT = 5
    FLUSH = 6
    FULL_HOUSE = 7
    FOUR_OF_A_KIND = 8
    STRAIGHT_FLUSH = 9

def getHands(hands, communityCards):
    for player in range(len(hands)):
        hands[player].extend(communityCards)   
    for player in range(len(hands)):
        for i in range(7):
            hands[player][i] = (hands[player][i][0], hands[player][i][1])
    return hands

def get5(hand):
    hand5 = []
    for i in combinations(hand, 5):
        hand5.append(i)
    print(f"hand5: {hand5}")
    return hand5

def isFourOfAKind(hand):
    ranks = []
    for card in hand:
        ranks.append(card[0])
    print(f"ranks: {ranks}")
    

def isFlush(hand):
    suits = []
    for card in hand:
        suits.append(card[1])
    print(f"suits: {suits}")
    if len(set(suits)) == 1:
        return True
    else:
        return False


def evaluateHand(hand):
    for hand in get5(hand):
        if isFlush(hand):
            print('flush')
```

```python
# take input as an array
holeCards = [['Kظآب', '3ظآخ'], ['9ظآب', '9ظآح'], ['9ظآث', 'Jظآث']]
communityCards = ['2ظآث', '5ظآث', '9ظآخ', '3ظآح', 'Kظآث']

playerHands = getHands(holeCards, communityCards)
evaluateHand(playerHands[2])
```
